##### import pandas as pd

```python
plant_data = pd.read_csv("status_from_USDA_MOSHER.csv")
```

```python
animal_data = pd.read_csv("status_from_AWI_MOSHER.csv")
```

```python
animal_data.set_index("Unnamed: 0")
```

```python
plant_data["Taxon Group"] = "Vascular Plant"
plant_data.head()
```

```python
plant_data.loc[plant_data.Status == "T", "Status"] = "Threatened"
```

```python
plant_data.loc[plant_data.Status == "E", "Status"] = "Endangered"
```

```python
plant_data = plant_data[["Common Name", "Scientific Name", "Status", "Taxon Group"]]
```

```python
plant_data.head()
```

```python
combined_dataframe = pd.concat([animal_data, plant_data])
```

```python
combined_dataframe
```

```python
combined_dataframe = combined_dataframe.drop("Unnamed: 0", 1)
```

```python
combined_dataframe
```

```python
combined_dataframe.to_csv("status_all_taxon_MOSHER.csv")
```
